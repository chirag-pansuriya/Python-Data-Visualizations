```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("titanic")
print(df.head(4))
```

```python
df.head(4)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='age',data=df)
plt.title("distribution of age titanic dataset")
plt.xlabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='who',y='age',data=df)
plt.title("distribution of age titanic dataset")
plt.xlabel("who")
plt.ylabel("age")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='who',y='age',data=df,hue='survived')
plt.title("distribution of age titanic dataset")
plt.xlabel("who")
plt.ylabel("age")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style("darkgrid")
sns.boxplot(x='who',y='age',data=df,hue='survived',palette='Set2')
sns.despine(trim=True)
plt.title("distribution of age titanic dataset")
plt.xlabel("who")
plt.ylabel("age")
plt.legend(title='SURVIVED')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style("darkgrid")
sns.boxplot(x='class',y='age',data=df,palette='Set2')
sns.swarmplot(x='class',y='age',data=df,size=3,color='gold',alpha=0.6)
sns.despine(trim=True)
plt.title("distribution of age titanic dataset")
plt.xlabel("who")
plt.ylabel("age")

plt.show()
```

```python
g=sns.FacetGrid(df,col='embark_town',height=5,aspect=1)
g.map(sns.boxplot,'class','age',order=['First','Second','Third'])
g.add_legend(title='Embark_town')
plt.show()
```

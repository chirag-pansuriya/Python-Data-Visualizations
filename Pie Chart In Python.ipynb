```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("titanic")
print(df.head(4))
```

```python
class_counts=df['class'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(class_counts,labels=class_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution Of Passengers')
plt.show()
```

```python
custom_colors=['gold','green','orange']
explode=(0,0,0.1)

class_counts=df['class'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(class_counts,labels=class_counts.index,autopct='%1.1f%%',startangle=90,explode=explode,colors=custom_colors)
plt.title('Distribution Of Passengers')
plt.show()
```

```python
custom_colors=['gold','green','orange']
explode=(0,0,0.1)

class_counts=df['who'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(class_counts,labels=class_counts.index,autopct='%1.1f%%',startangle=90,explode=explode,colors=custom_colors)
plt.title('Distribution Of Passengers')
plt.legend(class_counts.index,title='Passengers',loc='upper left')
plt.axis('equal')
plt.show()
```

```python
fig, ax=plt.subplots(1,2, figsize=(16,8))

ax[0].pie(class_counts,labels=class_counts.index,autopct='%1.1f%%',startangle=90,explode=explode,colors=custom_colors)
ax[0].set_title("Passengers distribution")
ax[0].axis('equal')

class_countss=df['survived'].value_counts()
ax[1].pie(class_countss,labels=['Survived','Not-Survived'],autopct='%1.1f%%',startangle=90,colors=['red','blue'])
ax[1].set_title("Survived/Not Survived")
ax[1].axis('equal')

plt.suptitle("Titanic Analysis")
plt.show()

```

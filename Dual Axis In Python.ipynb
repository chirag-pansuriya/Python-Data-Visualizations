```python
import pandas as pd
import numpy as np

np.random.seed(0)
data_range=pd.date_range(start='2022-01-01',periods=12,freq='M')
sales=np.random.randint(1000,5000,len(data_range))
profit=sales*np.random.uniform(0.1,0.3,len(data_range))

Sales_Data=pd.DataFrame({'Month':data_range,
                         'Sales':sales,
                         'Profit':profit})

Sales_Data.head(10)
```

```python
import matplotlib.pyplot as plt

fig,ax1=plt.subplots(figsize=(10,6))

ax1.plot(Sales_Data['Month'],Sales_Data['Sales'],label='sales',color='b')
ax1.set_xlabel('Month')
ax1.set_ylabel('sales')

ax2=ax1.twinx()
ax2.plot(Sales_Data['Month'],Sales_Data['Profit'],label='profit',color='g',linestyle='-')
ax2.set_xlabel('Month')
ax2.set_ylabel('profit')
plt.title("Dual axis chart for sales and profit")
plt.show()

```

```python
import matplotlib.pyplot as plt

fig,ax1=plt.subplots(figsize=(10,6))

ax1.plot(Sales_Data['Month'],Sales_Data['Sales'],label='sales',color='b',linestyle='-.',linewidth=5)
ax1.set_xlabel('Month')
ax1.set_ylabel('sales',color='b')
ax1.grid(True,linestyle='--',alpha=0.5)

ax2=ax1.twinx()
ax2.plot(Sales_Data['Month'],Sales_Data['Profit'],label='profit',color='g',linestyle='--',linewidth=2)
ax2.set_xlabel('Month')
ax2.set_ylabel('profit',color='g')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title("Dual axis chart for sales and profit")
plt.show()

```

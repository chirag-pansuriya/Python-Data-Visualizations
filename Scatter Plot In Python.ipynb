```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("titanic")
print(df.head(4))
```

```python
df.head(4)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age',y='fare',data=df)
plt.title("Age Vs Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age',y='fare',data=df)
plt.title("Age Vs Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.scatterplot(x='age',y='fare',hue='class',data=df)
plt.title("Age Vs Fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
custom_colors={'First':'red','Second':'gold','Third':'yellow'}
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.scatterplot(x='age',y='fare',hue='class',data=df,palette=custom_colors)
plt.title("Age Vs Fare",fontsize='15')
plt.xlabel('Age',fontsize='15')
plt.ylabel('Fare',fontsize='15')
plt.grid(True)
plt.show()
```

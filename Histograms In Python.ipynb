```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("titanic")
df.head(4)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='age')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of age groups")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.histplot(data=df,x='age',kde=True,hue="who",color='coolwarm',element='step')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of age groups")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='class', kde=True,  palette='dark',element='step')
plt.title('Age Distribution by Passenger Class ')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='class', kde=True,  palette='dark',element='step',bins=10,binrange=(0,80))
plt.title('Age Distribution by Passenger Class ')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
g=sns.FacetGrid(df,col='class',height=4,aspect=1)
g.map(sns.histplot,'age',kde=True,bins=10,color='gold')
g.set_titles('class:{col_name}')
g.set_axis_labels('age','count')
plt.show()
```

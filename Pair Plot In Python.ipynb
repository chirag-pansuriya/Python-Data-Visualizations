```python
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")

print(iris.head(4))
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.pairplot(iris,hue='species',palette='coolwarm',markers=['o','s','D'])
plt.show()
```

```python
sns.pairplot(iris,hue='species',diag_kind='kde',height=2.5)
plt.xlabel("Measurement")
plt.ylabel("Density")
plt.show()
```
